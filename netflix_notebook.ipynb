```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("/content/drive/MyDrive/DATA PROJECTS/DataCamp_proj/workspace (1)/workspace/netflix_data.csv")
```

The code below analyzes the dataset (netflix_df) of movies and shows available on Netflix. The goal is to filter and analyze movies that were released in the 1990s. The code performs the following key steps:

**Filter Movies Released in the 1990s:**
It first filters the dataset to include only those titles released between 1990 and 1999.

**Filter for Movies Only:**
It further filters the dataset to include only rows where the type is "Movie".
Plot a Histogram:

A histogram of the duration column (movie durations) is then created to visualize the distribution of movie lengths from the 1990s.



```python
# Start coding here! Use as many cells as you like
# Select all movies that was released in the 90s
All_release_90s = netflix_df[(netflix_df["release_year"]>= 1990 ) & (netflix_df["release_year"] <= 1999)]
#type(All_release_90s)
# filter out other types except the ones with the "type" = "Movie"
movies_90s = All_release_90s[All_release_90s["type"] == "Movie"]

duration_col = movies_90s["duration"]
plt.hist(duration_col, bins=20, edgecolor='black')
plt.title('Distibution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
#from the histogram, the most frequent movie duration is aproximately
duration = 100

```

From the histogram, What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called `duration`.

**Solution**:


```python
sol_str = 'Identifying the Most Frequent Movie Duration in the project: From the histogram, it is observed that the most frequent movie duration for all movies in the 1990s is approximately'
duration = 100
print(sol_str, duration, 'minutes.')
```

# How many short action movies are released in the '90s?
Movies with duration less than 90 minutes are generally considered to be short.
The following code identifies the count of short action movies released in the '90s:

```python
#Computing the count of short action movies
#Action movies in the 1990s
a_movies_90s = movies_90s[movies_90s['genre']=="Action"]
count = 0
for index,row in a_movies_90s.iterrows() : # iterate rows and check the duration column for values less than 90 and count
    if row['duration'] < 90 :
        count = count + 1
    else :
        count
short_movie_count = count
print("There are " + str(short_movie_count) + " short action movies in the 90s.")


```
